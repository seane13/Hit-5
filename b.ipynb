```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
years = list(range(2007, 2026))  # Update year range per site's data availability
all_records = []
```

```python
for year in years:
    url = f"https://walottery.com/WinningNumbers/PastDrawings.aspx?unittype=year&unitcount={year}&game=hit5"
    resp = requests.get(url)
    soup = BeautifulSoup(resp.text, 'html.parser')
    table = soup.find("table")
    if not table:
        continue
    for row in table.find_all('tr')[1:]:
        cols = [c.text.strip() for c in row.find_all('td')]
        if len(cols) >= 6:  # adjust if extra columns
            record = {
                "Date": cols[0],
                "Number1": cols[1],
                "Number2": cols[2],
                "Number3": cols[3],
                "Number4": cols[4],
                "Number5": cols[5]
            }
            all_records.append(record)

df = pd.DataFrame(all_records)
df.to_csv("hit5_historical_numbers.csv", index=False)
print("Scraping complete. Saved to hit5_historical_numbers.csv")
```
